I have done Sentiment Analysis for the following twitter accounts Cristiano Ronaldo , Mercedes Benz , Tim Cook and analyzed that these three accounts tweets are both positive and negative, but mostly positive. 
I have selected these accounts as they are from different background.ronaldo is from sports , Mercedes Benz is from automibile and Tim Cook is from business,Apple.
These accounts holders are happy and energetic people as they are having mostly positive tweets.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('Cristiano_tweets.pkl', 'rb') as f:
    Cristiano_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

CristianoX, CristianoY, CristianoZ = split_sentiments([get_sentiments(tweet.text) for tweet in Cristiano_tweets])
ax.scatter(CristianoX, CristianoY, CristianoZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('Cristiano')
plt.show()



from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('MercedesBenz_tweets.pkl', 'rb') as f:
    MercedesBenz_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

MercedesBenzX, MercedesBenzY, MercedesBenzZ = split_sentiments([get_sentiments(tweet.text) for tweet in MercedesBenz_tweets])
ax.scatter(MercedesBenzX, MercedesBenzY, MercedesBenzZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('MercedesBenz')
plt.show()



from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('tim_cook_tweets.pkl', 'rb') as f:
    tim_cook_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

tim_cookX, tim_cookY, tim_cookZ = split_sentiments([get_sentiments(tweet.text) for tweet in tim_cook_tweets])
ax.scatter(tim_cookX, tim_cookY, tim_cookZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('tim_cook')
plt.show()



plt.scatter(xs,zs,label='Cristiano')
plt.scatter(MercedesBenzX,MercedesBenzZ,color='r',marker='*',label='MercedesBenz')
plt.scatter(tim_cookX,tim_cookZ,color='y',marker='d',label='tim_cook')
plt.xlabel('negative')
plt.ylabel('positive')
plt.title('positive vs negative Sentiments')
plt.legend(loc='upper right',shadow=True)
plt.show()
```
